```python
import pandas as pd

# Define a function to flatten the nested structure
def flatten_structure(chapter):
    data = []
    
    chapter_number = chapter["Chapter Number"]
    chapter_title = chapter["Chapter Title"]
    chapter_summary = chapter["Chapter Summary"]
    
    for section in chapter["Sections"]:
        section_number = section["Section Number"]
        section_title = section["Section Title"]
        section_summary = section["Section Summary"]
        
        for subsection in section.get("Subsections", []):
            subsection_number = subsection["Subsection Number"]
            subsection_title = subsection["Subsection Title"]
            subsection_summary = subsection["Subsection Summary"]
            
            for heading in subsection.get("Headings", []):
                heading_number = heading["Heading Number"]
                heading_title = heading["Heading Title"]
                heading_summary = heading["Heading Summary"]
                
                for subheading in heading.get("Subheadings", []):
                    subheading_number = subheading["Subheading Number"]
                    subheading_title = subheading["Subheading Title"]
                    subheading_summary = subheading["Subheading Summary"]
                    
                    content = subheading.get("Content", "")
                    
                    data.append([
                        chapter_number, chapter_title, chapter_summary,
                        section_number, section_title, section_summary,
                        subsection_number, subsection_title, subsection_summary,
                        heading_number, heading_title, heading_summary,
                        subheading_number, subheading_title, subheading_summary,
                        content
                    ])
    
    return data

# Initialize an empty list to collect the flattened data
flattened_data = []

# Iterate through the chapters and flatten the structure
for chapter in chapters:
    flattened_data.extend(flatten_structure(chapter))

# Create a DataFrame with appropriate column names
df = pd.DataFrame(flattened_data, columns=[
    "Chapter Number", "Chapter Title", "Chapter Summary",
    "Section Number", "Section Title", "Section Summary",
    "Subsection Number", "Subsection Title", "Subsection Summary",
    "Heading Number", "Heading Title", "Heading Summary",
    "Subheading Number", "Subheading Title", "Subheading Summary",
    "Content"
])

# Display the DataFrame
df
```
