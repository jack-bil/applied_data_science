```python
import random

# Define the number of possible actions (guesses)
num_actions = 5

# Define the Q-table as a dictionary with state-action pairs
Q = {i: [0] * num_actions for i in range(1, num_actions + 1)}

# Hyperparameters
learning_rate = 0.5
discount_factor = 0.9
exploration_prob = 0.2
num_episodes = 10000

# Training the agent
for episode in range(num_episodes):
    # Randomly select the target number (the goal)
    target_number = 1
    
    # Reset the environment (initial state is the target number)
    state = target_number
    done = False

    while not done:
        # Choose an action (guess) using epsilon-greedy policy
        if random.random() < exploration_prob:
            action = random.randint(1, num_actions)  # Explore (random guess)
        else:
            action = Q[state].index(max(Q[state])) + 1  # Exploit (choose best guess)

        # Check if the guessed number matches the target number
        if action == target_number:
            reward = 1
            done = True
        else:
            reward = 0

        # Update the Q-table using the Q-learning formula
        Q[state][action - 1] = (1 - learning_rate) * Q[state][action - 1] + \
                               learning_rate * (reward + discount_factor * max(Q[state]))

        # Move to the next state (the guessed number)
        state = action

# Test the learned policy
test_episodes = 10000
for _ in range(test_episodes):
    target_number = random.randint(1, num_actions)
    state = target_number
    done = False

    while not done:
        action = Q[state].index(max(Q[state])) + 1
        if action == target_number:
            print(f"Target number: {target_number}, Agent's guess: {action} (Correct)")
            done = True
        else:
            print(f"Target number: {target_number}, Agent's guess: {action} (Incorrect)")
            done = True

```
